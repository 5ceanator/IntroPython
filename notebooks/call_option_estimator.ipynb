```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/Users/keemsunguk/projects/intro/data/tr_eikon_eod_data.csv', index_col=0, parse_dates=True)
```

```python
data_spx = pd.DataFrame(data['.SPX'])
```

```python
data_spx['rets'] = np.log(data_spx/data_spx.shift(1))
data_spx['vola'] = data_spx['rets'].rolling(252).std() * np.sqrt(252)
data_spx[['.SPX', 'vola']].plot(subplots=True, figsize=(10, 6));
plt.savefig('/Users/keemsunguk/projects/intro/images/spx_volatility.png')
```
